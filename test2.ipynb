### *1. Enregistrement audio*

```python
import sounddevice as sd
from scipy.io.wavfile import write
from faster_whisper import WhisperModel

fs = 44100
seconds = 5

print("Parlez..")

audio = sd.rec(
    int(seconds * fs),
    samplerate=fs,
    channels=1
)

sd.wait()
write("audio.wav",fs, audio)

print("ordre reçu".center(100,'='))
```

### *2. Instanciation du modèle et conversion de l'audio en txt* 

```python
print("Traitement....")

model = WhisperModel(
    "large-v1",
    device="cpu",
    compute_type="int8"
)

segments,info = model.transcribe(
    "audio.wav",
    language="fr"
)

text= ""

for i in segments:
    text += i.text

text
```

### *3. Attribution du rôle au llm et traitement de la commande*

```python
import ollama

SYSTEM_PROMPT = open("SYS_PROMPT.txt","r",encoding="UTF-8").read()

reponse = ollama.chat(
    "llama3:latest",
    messages=[
        {
            'role' : "system",
            'content' : SYSTEM_PROMPT
        },
        {
            'role' : "user",
            'content' : text
        }
    ],
    stream=True,
    #think="medium"

)
rps = ""
for i in reponse:
    rps += i["message"]["content"]
```

```python
rps = rps.replace("'","").replace("TOI","")
rps
```
